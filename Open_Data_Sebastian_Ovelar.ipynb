En objeto_causa existen valores distintos que se podrian unificar a uno, los cuales corrijo reemplanzandolos por el correcto

```python
df['objeto_causa'] = df['objeto_causa'].replace('Violación - intimación**', 'Violación - intimación')
df['objeto_causa'] = df['objeto_causa'].replace('Daños a cosas de interes común**', 'Daños a cosas de interes común')
df['objeto_causa'] = df['objeto_causa'].replace('Homicidio**', 'Homicidio')
df['objeto_causa'] = df['objeto_causa'].replace('Homicidio doloso', 'Homicidio Doloso')
df['objeto_causa'] = df['objeto_causa'].replace('Homicidio Doloso**', 'Homicidio Doloso')
df['objeto_causa'] = df['objeto_causa'].replace('Abigeato**', 'Abigeato')
df['objeto_causa'] = df['objeto_causa'].replace('Abuso sexual niños - estímulo sexual**', 'Abuso sexual niños - estímulo sexual')
df['objeto_causa'] = df['objeto_causa'].replace('Coacción - Sexual', 'Coacción Sexual')
df['objeto_causa'] = df['objeto_causa'].replace('Contaminación del aire y ruidos capaces de dañar la salud**', 'Contaminación del aire y ruidos capaces de dañar la salud')
df['objeto_causa'] = df['objeto_causa'].replace('Delitos contra el medio ambiente**', 'Delitos contra el medio ambiente')
df['objeto_causa'] = df['objeto_causa'].replace('Hurto - Tentativa**', 'Hurto - Tentativa')
df['objeto_causa'] = df['objeto_causa'].replace('Incumplimiento del deber legal alimentario, establecido en un convenio judicialmente aprobado o en una resolución judicial**', 'Incumplimiento del deber legal alimentario, establecido en un convenio judicialmente aprobado o en una resolución judicial')
df['objeto_causa'] = df['objeto_causa'].replace('Intervenciones peligrosas en el tránsito terrestre**', 'Intervenciones peligrosas en el tránsito terrestre')
df['objeto_causa'] = df['objeto_causa'].replace('Lesión culposa (Daño a la Salud)', 'Lesión culposa')
df['objeto_causa'] = df['objeto_causa'].replace('Manipulación de graficaiones técnicas', 'Manipulación de graficaciones técnicas')
df['objeto_causa'] = df['objeto_causa'].replace('Obstrucción al resarcimiento por daños en accidentes de tran', 'Obstrucción al resarcimiento por daños en accidentes de tránsito')
df['objeto_causa'] = df['objeto_causa'].replace('Tráfico de drogas. Tenencia, consumo y medidas de seguridad curativa**', 'Tráfico de drogas. Tenencia, consumo y medidas de seguridad curativa')
df['objeto_causa'] = df['objeto_causa'].replace('Tráfico de drogas. tráfico y delitos conexos**', 'Tráfico de drogas. tráfico y delitos conexos')
df['objeto_causa'] = df['objeto_causa'].replace('Tráfico Ilicito de armas de fuego(Delitos Econ.)', 'Tráfico Ilicito de armas de fuego')
df['objeto_causa'] = df['objeto_causa'].replace('Hurto agravado', 'Hurto Agravado')
```

Este codigo es para comprobar si el cambio se ha hecho correcto, sumando los datos que les pida que me brinde

```python
conteo = (df['objeto_causa'] == 'Hurto Agravado').sum()
print(conteo)
```
